```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x, y = np.loadtxt('video-short-sched.data', delimiter=' ', usecols=(0, 1), unpack=True)
xo, yo = np.loadtxt('video-short-ondemand.data', delimiter=' ', usecols=(0, 1), unpack=True)

fig = plt.figure()

ax1 = fig.add_subplot(111)

ax1.set_title("video-short.json")    
ax1.set_xlabel('time (s)')
ax1.set_ylabel('Energy (mJ)')

ax1.plot(x,y, c='r', label='sched')
ax1.plot(xo,yo, label='ondemand')

leg = ax1.legend()
```
